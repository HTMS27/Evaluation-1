```python
from flask import Flask, render_template, request,make_response
from werkzeug.utils import secure_filename
import pandas as pd
from transformers import pipeline
import re

app = Flask(__name__)


        
@app.route('/')
def upload_file():
    return render_template('upload.html')



@app.route('/uploader', methods = ['GET', 'POST'])
def uploader():
    if request.method == 'POST':
        f = request.files['file']
        f.save(secure_filename(f.filename))
        data=pd.read_csv(secure_filename(f.filename))
        sentiment_analysis=pipeline("sentiment-analysis")
        data['Text'] = data['Text'].str.replace('[^\w\s#@/:%.,_-]', '',flags=re.UNICODE)
        data_star=data[['Star','Text']]
        data_low_star=data_star[data_star['Star']<=2]
        data_low_star['sentiment']=data_low_star['Text'].apply(lambda x:sentiment_analysis(x)[0]['label'])
        data_low_star['sentiment_score']=data_low_star['Text'].apply(lambda x:sentiment_analysis(x)[0]['score'])
        data_sentiment_pos=data_low_star[data_low_star['sentiment']=='POSITIVE']
        st=''
        for (i,j) in zip(data_low_star[(data_low_star['sentiment']=='POSITIVE') & (data_low_star['sentiment_score']>0.999)]['Text'],data_low_star[(data_low_star['sentiment']=='POSITIVE') & (data_low_star['sentiment_score']>0.999)]['Star']):
            st=st+str(str(i)+'-'+str(j))+'\t\t'
        
        return str(data.columns)
    
        
if __name__ == '__main__':
    app.run()
```
